```python

import os
import pandas as pd
import random as rn
import numpy as np
from collections import Counter
from IPython.display import display, Image
import tensorflow as tf
```

```python
# 딕셔너리화
image_feature_description = {
    'height': tf.io.FixedLenFeature([], tf.int64),
    'width': tf.io.FixedLenFeature([], tf.int64),
    'depth': tf.io.FixedLenFeature([], tf.int64),
    'label': tf.io.FixedLenFeature([24], tf.int64), 
    'image_raw': tf.io.FixedLenFeature([], tf.string),
}


# 사전을 이용하여 입력 tf.Example 프로토를 구문 분석
def _parse_image_function(example_proto):
    return tf.io.parse_single_example(example_proto, image_feature_description)

# tfrecord 파일을 불러옴
raw_image_dataset = tf.data.TFRecordDataset('/content/drive/MyDrive/facial_emotion_recognition_using_K-drama_dataset-main/tfrecords/multi-label_test.tfrecords')
# _parse_image_function을 이용한 raw_image_dataset 값 변경
parsed_image_dataset = raw_image_dataset.map(_parse_image_function)
parsed_image_dataset
```

```python
n = 0
for image_features in parsed_image_dataset.take(3882):
    image_raw = image_features['image_raw'].numpy()
    # display(Image(data=image_raw))
    n += 1
    print("Number of images displayed:", n,"===>", image_features['label'])
```
